```python
# python program to print prime factors
import numpy as np
import math
def primefactors(n):
    divisori = []
    
    #even number divisible
    while n % 2 == 0:
        divisori.append(2)
        #print (2),
        n = n / 2

    #n became odd
    for i in range(3,int(math.sqrt(n))+1,2):

        while (n % i == 0):
#            print (i)
            divisori.append(i)
            n = n / i

    if n > 2:
#        print (n)
        divisori.append(n)

    return len(divisori)

 
#n = int(input("Enter the number for calculating the prime factors :\n"))
#primefactors(n)
```

```python
primefactors(5)
```

```python
my_hist = []
for i in np.arange(1,1000000):
    my_hist.append(primefactors(i))

from matplotlib import pyplot as plt
plt.hist(my_hist, bins= max(my_hist)-min(my_hist))
```
